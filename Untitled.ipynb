```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("India_Injury_Road_Accident_Fatality_2017-2020.csv")
df.head()
```

```python
df.info()
df.describe()
print(df.columns)
```

```python
# Step 1: Clean the column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
#print(df.columns.tolist())
df_sorted = df.sort_values(by='Road Accidents during 2018', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Road Accidents during 2018', y='State',hue='State', data=df_sorted, palette='coolwarm')
plt.title('Road Accidents by State in 2018')
plt.xlabel('Number of Accidents')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Road Accidents during 2019', y='State',hue='State', data=df_sorted, palette='coolwarm')
plt.title('Road Accidents by State in 2019')
plt.xlabel('Number of Accidents')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Road Accidents during 2020', y='State',hue='State', data=df_sorted, palette='coolwarm')
plt.title('Road Accidents by State in 2020')
plt.xlabel('Number of Accidents')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
df_stacked = df[['State','Accidents per 10,000 Vehicles - 2017', 
                 'Accidents per 10,000 Vehicles - 2018', 
                 'Accidents per 10,000 Vehicles - 2019']]
df_stacked.set_index('State',inplace=True)

df_stacked.plot(kind='bar',stacked=True,figsize=(14,8),colormap='viridis')

plt.title('Accidents per 10,000 Vehicles (2017â€“2019) by State')
plt.xlabel('State')
plt.ylabel('Accidents per 10,000 Vehicles')
plt.legend(title='year')
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
df_stacked = df[['State','Accidents per 10,000 Km of Roads - 2017', 
                 'Accidents per 10,000 Km of Roads - 2018']]
df_stacked.set_index('State',inplace=True)

df_stacked.plot(kind='bar',stacked=True,figsize=(14,8),colormap='viridis')

plt.title('Accidents per 10,000 km (2017â€“2019) by State')
plt.xlabel('State')
plt.ylabel('Accidents per 10,000 km')
plt.legend(title='year')
plt.tight_layout()
plt.show()
```
